```python
def countDict(array):
    # Создадим словарь для отслеживания уникальных вторых элементов
    second_elements = {}

    # Пройдемся по каждому элементу массива и обновим словарь
    for sublist in array:
        if sublist[1] not in second_elements:

            second_elements[sublist[1]] = 1
        else:
            second_elements[sublist[1]] += 1

    printDict(second_elements)

    return second_elements

def printDict(d):
    # Выведем результаты
    for key, value in d.items():
        
        print(f"{key}: {value}")


def countMax(regression_la):

    second_elements = countDict(regression_la)

    # Находим максимальное значение в словаре
    max_count = max(second_elements.values())

    # Находим ключ(и) с максимальным значением
    most_common_elements = [key for key, value in second_elements.items() if value == max_count]


    print("Наиболее часто встречающийся элемент(ы):", most_common_elements)
    print("Количество вхождений:", max_count)

    return most_common_elements[0]



def filter_arrays(arr):
    mx = countMax(arr)
    return [sub_arr for sub_arr in arr if sub_arr[1] == mx]



# Фильтрация массива
filtered_regression_la = filter_arrays(regression_la)

print("Исходный массив:", regression_la)
print("Отфильтрованный массив:", filtered_regression_la)


```

## Корреляционный анализ
Применим метод скользящего окна, который будет перемещать окно заданной длины по массиву seq2 и вычислять корреляцию с каждой подпоследовательностью из seq1. 

### Корреляция Пирсона: 
используется для измерения корреляции между двумя непрерывными переменными. 
### Корреляция Спирмена: 
используется для измерения корреляции между двумя ранжированными переменными.
### Корреляция Кендалла: 
используется, когда вы хотите использовать корреляцию Спирмена, но размер выборки мал и имеется много связанных рангов.

```python
from scipy.stats import pearsonr, spearmanr, kendalltau
import matplotlib.pyplot as plt

def find_best_correlation(seq1, seq2, sub_seq_length, correlation_type):
    correlation_functions = {
        'pearson': pearsonr,
        'spearman': spearmanr,
        'kendall': kendalltau
    }

    correlation_function = correlation_functions.get(correlation_type.lower())
    if correlation_function is None:
        raise ValueError("такого нет( выберите 'pearson', 'spearman' или 'kendall'.")

    best_correlation = None
    best_sequence_indices = None

    for i in range(len(seq2) - sub_seq_length + 1):
        sub_seq2 = seq2[i:i + sub_seq_length]
        for j in range(len(seq1) - sub_seq_length + 1):
            sub_seq1 = seq1[j:j + sub_seq_length]
            correlation, _ = correlation_function(sub_seq1, sub_seq2)
            if best_correlation is None or correlation > best_correlation:
                best_correlation = correlation
                best_sequence_indices = (i, i + sub_seq_length), (j, j + sub_seq_length)

    return best_correlation, best_sequence_indices



# Размер подпоследовательности для поиска
sub_seq_length = 50

# Тип корреляции: 'pearson', 'spearman' или 'kendall'
correlation_type = 'spearman'

# Найти наилучшую корреляцию и соответствующую подпоследовательность
best_correlation, best_indices = find_best_correlation(regression_kip_values, filtered_regression_la_values, sub_seq_length, correlation_type)

if best_indices is not None:
    best_sub_seq1 = regression_kip_values[best_indices[1][0]:best_indices[1][1]]
    best_sub_seq2 = filtered_regression_la_values[best_indices[0][0]:best_indices[0][1]]
    print("Наилучшая корреляция:", best_correlation)
    print("Соответствующая подпоследовательность regression_kip_values:", best_sub_seq1)
    print("Соответствующая подпоследовательность filtered_regression_la_values:", best_sub_seq2)
    # Визуализация точек на плоскости
    #plt.scatter(regression_kip_values, filtered_regression_la_values, color='blue', label='All Points')
    plt.scatter(best_sub_seq1, best_sub_seq2, color='blue')
    plt.xlabel('regression_kip_values')
    plt.ylabel('filtered_regression_la_values')
    plt.title("Корреляция Спирмена")
    plt.legend()
    #plt.show()
    plt.savefig(correlation_type)
    
else:
    print("Подпоследовательность не найдена.")
```

```python
from datetime import datetime

# Преобразование строк в даты и числа
def convert_data(data):
    converted_data = []
    for sublist in data:


        new_sublist = []
        for item in sublist:

            if len(item) == 3:
                # Преобразование даты и времени
                date_time = datetime.strptime(f"{item[0]} {item[1]}", "%d.%m.%Y %H:%M")
                # Преобразование числа
                value = float(item[2])
                new_sublist.append([date_time, value])
        converted_data.append(new_sublist)
    return converted_data

# Функця вывода данных 
def print_data(data):
    for index, sublist in enumerate(data):
        print(f"Элемент {index + 1}:")
        for item in sublist:
            print(f"Дата: {item[0].strftime('%d.%m.%Y %H:%M')}, Значение: {item[1]}")
        print()



# Преобразование данных
result_regression_kip_converted = convert_data(result_regression_kip)
result_regression_la_converted = convert_data(result_regression_la)


print_data(result_regression_kip_converted)
print(result_regression_kip_converted)

print("\n")

print_data(result_regression_la_converted)

print(result_regression_la_converted)



```

```python
# Преобразование кубических массивов в словари, где ключами будут даты
kip_dict = {item[0]: item[1] for sublist in result_regression_kip_converted for item in sublist}
#print(kip_dict)
la_dict = {item[0]: item[1] for sublist in result_regression_la_converted for item in sublist}
#print(la_dict )

# Создание словаря для разниц во времени
time_diff_dict = {}

for kip_time, kip_value in kip_dict.items():
    for la_time, la_value in la_dict.items():
        time_diff = abs(kip_time - la_time)
        time_diff_dict[(kip_time, la_time)] = {'kip_value': kip_value, 'la_value': la_value, 'time_diff': time_diff}

# Вывод словаря 
for key, value in time_diff_dict.items():
    print(f"Время KIP: {key[0]}, значение KIP: {value['kip_value']}")
    print(f"Время LA: {key[1]}, значение LA: {value['la_value']}")
    print(f"Разница времени: {value['time_diff']}")
    print()

```

```python
def print_min_time_diff(time_diff_dict):
    min_diff_days = min(value['time_diff'].days for value in time_diff_dict.values())
    
    print("Элементы с минимальной разницей в днях:")
    for key, value in time_diff_dict.items():
        if value['time_diff'].days == min_diff_days:
            print(f"Время KIP: {key[0]}, значение KIP: {value['kip_value']}")
            print(f"Время LA: {key[1]}, значение LA: {value['la_value']}")
            print(f"Разница времени: {value['time_diff']}")
            print()

print_min_time_diff(time_diff_dict)
```

```python
import pandas as pd

df = pd.read_excel('Смешение.xlsx', header=None)
mixing = []
n = 5  # кол-во читаемых строк
for index, row in df.iterrows():
    tmp = ','.join(map(str, row)).split(",")
    mixing.append(tmp[0].split(" "))
    tmp.pop(0)
    for el in tmp:
        mixing[index].append(el)
            
    if index == n-1:
        break
    
mixing.pop(0)
print(mixing)
```

```python

import numpy as np
from scipy.optimize import curve_fit
import datetime

# Извлечение значений Q1(t), Q2(t), Q3(t) и t из mixing
t = [entry[0] for entry in mixing]
Q1 = [entry[2] for entry in mixing]
Q2 = [entry[3] for entry in mixing]
Q3 = [entry[4] for entry in mixing]

# Преобразование времени в числовой формат (например, секунды с начала эпохи Unix)
t_numeric = [(entry[0] - datetime.datetime(1970, 1, 1)).total_seconds() for entry in mixing]

# Функция для линейной зависимости Q(t) = a * t + b
def linear_func(t, a, b):
    return a * t + b

# Фитирование данных для Q1(t)
params_Q1, _ = curve_fit(linear_func, t_numeric, Q1)
a_Q1, b_Q1 = params_Q1

# Фитирование данных для Q2(t)
params_Q2, _ = curve_fit(linear_func, t_numeric, Q2)
a_Q2, b_Q2 = params_Q2

# Фитирование данных для Q3(t)
params_Q3, _ = curve_fit(linear_func, t_numeric, Q3)
a_Q3, b_Q3 = params_Q3

print("Для Q1(t):")
print("a =", a_Q1)
print("b =", b_Q1)
print("\nДля Q2(t):")
print("a =", a_Q2)
print("b =", b_Q2)
print("\nДля Q3(t):")
print("a =", a_Q3)
print("b =", b_Q3)


# Функция для Q1(t)
def Q1_func(dt):
    t = (dt - datetime.datetime(1970, 1, 1)).total_seconds()
    return a_Q1 * t + b_Q1

# Функция для Q2(t)
def Q2_func(dt):
    t = (dt - datetime.datetime(1970, 1, 1)).total_seconds()
    return a_Q2 * t + b_Q2

# Функция для Q3(t)
def Q3_func(dt):
    t = (dt - datetime.datetime(1970, 1, 1)).total_seconds()
    return a_Q3 * t + b_Q3

# Пример времени (год, месяц, день, час, минута, секунда)
example_time = datetime.datetime(2022, 6, 1, 0, 0, 0)

# Получение значений Q1(t), Q2(t), Q3(t) для данного времени
Q1_value = Q1_func(example_time)
Q2_value = Q2_func(example_time)
Q3_value = Q3_func(example_time)

print("Для времени", example_time, ":")
print("Q1(t) =", Q1_value)
print("Q2(t) =", Q2_value)
print("Q3(t) =", Q3_value)
```
